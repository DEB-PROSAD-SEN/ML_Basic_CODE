```python
#importing mathplotlib library
import matplotlib.pyplot as plt
```

import numpy to get data for our plots

```python
import numpy as np
x=np.linspace(0,10,100)
y=np.sin(x)
z=np.cos(x)
print(x)
print(y)
print(z)
```

plotting the data

```python
#for sin wave
plt.plot(x,y)
plt.show()
```

```python
#plot for cosin wave
plt.plot(x,z)
plt.show()
```

```python
#adding title x-axis & y-axis
plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sin value')
plt.title('sine wave')
plt.show()
```

plot a perabola

```python
u=np.linspace(-10,10,20)
d=u**2
plt.plot(u,d,'y.')
plt.show()
```

```python
v=np.linspace(-5,5,100)
plt.plot(v,np.sin(v),'g*')
plt.plot(v,np.cos(v),'r>')
```

bar plot

```python
fig = plt.figure()
ax=fig.add_axes([0,0,1,1])
languages = ["English","French","Spanish","Latin","German"]
people = [100,50,150,40,70]
ax.bar(languages,people)
plt.xlabel('Languages')
plt.ylabel('people')
plt.show()
```

pai chart

```python
fgl=plt.figure()
ad=fgl.add_axes([0,0,1,1])
languages = ["English","French","Spanish","Latin","German"]
people = [100,50,150,40,70]
ad.pie(people, labels=languages,autopct ='%1.1f%%')
fig2=plt.figure()
ay = fig2.add_axes([0,0,1,1])
ay.scatter(x,y,color='g')
ay.scatter(x,z,color='b')
plt.show()
```

3d scatter plot

```python
fig3=plt.figure()
ac=plt.axes(projection='3d')
r=20*np.random.random(100)
t=np.sin(r)
u=np.cos(r)
ac.scatter(t,u,r,c=r,cmap='Blues')
plt.show()
```
